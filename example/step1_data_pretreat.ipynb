```python
import os

import umap
from matplotlib.colors import LinearSegmentedColormap
from sklearn.decomposition import PCA
import scanpy as sc
GPU_NUMBER = [1,2,3,4,5]
os.environ["CUDA_VISIBLE_DEVICES"] = ",".join([str(s) for s in GPU_NUMBER])
import matplotlib.pyplot as plt
import datasets
from collections import Counter
from tqdm import tqdm
from geneformer import EmbExtractor
import pickle
import numpy as np
import igraph as ig
import leidenalg as la
from sklearn.neighbors import NearestNeighbors

def leiden_clustering(vectors, n_neighbors=10, resolution=1.0):
    """
    使用 Leiden 算法对输入的向量列表进行聚类。

    参数:
    vectors (list of list of float): 输入的向量列表。
    n_neighbors (int): 近邻数量。
    resolution (float): 聚类的分辨率。

    返回:
    list of int: 每个向量对应的聚类标签。
    """
    # 计算近邻图
    neighbors = NearestNeighbors(n_neighbors=n_neighbors, metric='euclidean')
    neighbors.fit(vectors)
    knn_graph = neighbors.kneighbors_graph(vectors, mode='connectivity')

    # 转换为 igraph 图
    sources, targets = knn_graph.nonzero()
    graph = ig.Graph(directed=False)
    graph.add_vertices(knn_graph.shape[0])
    edges = list(zip(sources, targets))
    graph.add_edges(edges)

    # 使用 Leiden 算法进行聚类
    partition = la.find_partition(graph, la.RBConfigurationVertexPartition, resolution_parameter=resolution)

    return partition.membership
def getEbeddings(dataset_path,dataset_target_path,target_path):
    #  获取所有标签细胞的embeddings，使用各位平均值与标签作图
    with open('../dataSets/gene_tk.pkl', 'rb') as f:
        gene_tk = pickle.load(f)
    cell_ebeddings = {}
    labels=[]
    data_all1 = []
    label = 'label'
    data_all2 = datasets.Dataset.load_from_disk(
        dataset_path)
    print('data_reload..')
    for item in tqdm(data_all2):
        item['emb_label'] = label + str(item['label'])
        if(label + str(item['label']) not in labels):
            labels.append(label + str(item['label']))
        data_all1.append(item)
    print(Counter(data_all2['label']))
    data_all1 = datasets.Dataset.from_list(data_all1)
    data_all1.save_to_disk(
        dataset_target_path)
    filter_dict = {}
    for k in tqdm(labels):
        filter_dict['emb_label'] = k
        embex = EmbExtractor(model_type="Pretrained",
                             filter_data=filter_dict,
                             emb_mode="cell",
                             # emb_layer=-1,
                             emb_layer=0,
                             max_ncells=50000,
                             forward_batch_size=12,
                             nproc=16,
                             token_dictionary_file='../dataSets/gene_tk.pkl')
        embs = embex.extract_embs(
            '../models/time_finetune',
            dataset_target_path,
            "../result/",
            "output_prefix")
        with open("../result/output_prefix.csv", 'r') as f:
            data_emb = f.readlines()
            for i in range(1, len(data_emb)):
                data_list = data_emb[i]
                data_list = data_list.strip('\n')
                data_list = data_list.split(',')
                if(k not in cell_ebeddings):
                    cell_ebeddings[k]=[[float(item) for item in data_list[1:]]]
                else:
                    cell_ebeddings[k].append([float(item) for item in data_list[1:]])

    with open(target_path, 'wb') as f:
        pickle.dump(cell_ebeddings, f)
    print(len(cell_ebeddings))
    print(cell_ebeddings.keys())
# GRN link dataset
def edge_class(edge_weight):
    if (edge_weight<-100):
        return 0
    if (edge_weight>=-100 and edge_weight<-15):
        return 1
    if (edge_weight>=-15 and edge_weight<-3):
        return 2
    if (edge_weight>=-3 and edge_weight<=3):
        return 3
    if (edge_weight>3 and edge_weight<15):
        return 4
    if (edge_weight>=15 and edge_weight<100):
        return 5
    if (edge_weight>=100):
        return 6
def edge_target():
    # 此方法用于生成图
    edge_pure = []
    # edge_classes=[]
    with open('../dataSets/edges.csv') as f:
        edges = f.readlines()
        for edge in edges[1:]:
            edge = edge.strip('\n')
            edge = edge.split(',')
            edge_pure.append([edge[1], edge[2], edge_class(float(edge[3]))])
    return edge_pure
    pass
```
